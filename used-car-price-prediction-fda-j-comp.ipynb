# Label Encoding
### Our dataset has 12 characteristic categorical variables and 4 numerical variables (not including the price column). To apply ML models, we need to convert these categorical variables into numerical variables and LabelEncoder software from sklearn library is used to solve this problem. The below figure depicts label encoding which is performed on the dataset.

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
X=['manufacturer','fuel','transmission','type','drive','condition',
   'cylinders','title_status'
  ,'state','paint_color','size']
   #,'state','paint_color','size']
for x in X:
    dataframe[x] = le.fit_transform(dataframe[x])
```

## Multiple Linear regression
### Multiple Linear regression model is the quickest and easiest regression model but the precision is lower than other models and used only in certain problems. It uses weights to fit the training data in a straight-line and weights are adjusted by minimizing the mean squared error loss.

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)
train_test_model(model,"Linear Regression",X_train,y_train,X_test,y_test)
```

## Ridge Regression
### Ridge regression is a regularized version of linear regressor but has alpha parameter to control the regulation of the model to reduce the variance of the estimates.

```python
from sklearn.linear_model import Ridge
model = Ridge(alpha=1,fit_intercept=True,random_state=1)
train_test_model(model,"Ridge Regression",X_train,y_train,X_test,y_test)
```

## LASSO Regression
### LASSO stands for Least Absolute Shrinkage and Selection Operator which is a shrinkage regularization technique. It is a simple sparse model which has only fewer parameters than ridge or linear regression. This type of regression is used when the dataset shows high multi collinearity or when you want to automate variable elimination and feature selection.


```python
from sklearn.linear_model import Lasso

model = Lasso(fit_intercept=True,random_state=1)
train_test_model(model,"Lasso Regression",X_train,y_train,X_test,y_test)
```

## Bayesian Regression
### Bayesian regression is a type of linear regression which can be implemented with insufficient data or poorly distributed data using probability distributors rather than point estimates.

```python
from sklearn.linear_model import BayesianRidge
model=BayesianRidge(fit_intercept=True)

train_test_model(model,"Bayesian Regression",X_train,y_train,X_test,y_test)
```

## Multi-layer Perceptron 
### Multi-layer Perceptron (MLP) Regression Algorithm is a type of supervised learning which emulate a function from the input dataset. It is an artificial neural network composed of multiple layers of perceptron. It uses hidden layers and neurons to predict the value of the unknown input.

```python
from sklearn.neural_network import MLPRegressor
model=MLPRegressor(hidden_layer_sizes=(28,28,42,1),activation='relu',learning_rate='adaptive', learning_rate_init=0.1, solver='adam', max_iter=320,random_state=1,verbose=True,early_stopping=True,n_iter_no_change=70)
train_test_model(model,"MLP Regression",X_train,y_train,X_test,y_test)
```
